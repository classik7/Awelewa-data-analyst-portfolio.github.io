```python
stat, p_value = shapiro(seun_data['Consistency'])
print('P value: ', round(p_value, 2))
#normality test
```

```python
Spain = seun_data.query('Nationality == "Spain"')['Consistency']
France = seun_data.query('Nationality == "France"')['Consistency']
Germany = seun_data.query('Nationality == "Germany"')['Consistency']
Brazil = seun_data.query('Nationality == "Brazil"')['Consistency']
England = seun_data.query('Nationality == "England"')['Consistency']
#filter test
```

```python
stats, p_value = levene(Spain, France, Germany, Brazil, England)
print("P value: ", round(p_value, 3))       #levene test
```

```python
Test_stat, p_value = f_oneway(Spain, France, Germany, Brazil, England)
print("P value: ", round(p_value, 2))
# one_way anova
```

```python
# Kmean clustering
x = df.drop('Striker_ID', axis = 1)

wcss = []

for i in range(1, 15):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(x)
    wcss_score = kmeans.inertia_
    wcss.append(wcss_score)
```

```python
plt.figure(figsize = (12, 6))
plt.plot(range(1, 15), wcss, marker = 'o')
plt.title('Elbow methods')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.show()

```

```python
# Building KMeans with k = 2
last_km = KMeans(n_clusters = 2)
last_km.fit_predict(x)

# Generating labels
labels = last_km.labels_
labels
```

```python
# Adding labels
df['Clusters'] = labels
df.head()
```
